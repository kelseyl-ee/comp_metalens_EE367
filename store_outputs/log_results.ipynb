```python
import csv
import os

folder_path = "." 
filenames = [
    "MNIST_class_results.csv", 
    "Fashion_class_results.csv", 
    "CIFAR_G_class_results.csv"
]
headers = ["Method", "Num_Kernels", "Kernel_Size", "Test Accuracy"]

for filename in filenames:
    # This will now create the files in the current directory
    file_path = os.path.join(folder_path, filename)
    
    if not os.path.exists(file_path):
        with open(file_path, mode='w', newline='') as f:
            writer = csv.writer(f)
            writer.writerow(headers)
        print(f"Created: {filename}")
    else:
        print(f"Skipped: {filename} (already exists)")
```
